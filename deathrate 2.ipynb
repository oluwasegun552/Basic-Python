#### Trend analysis

AFGHANISTAN DUE TO SMOKING

```python
import matplotlib.pyplot as plt
```

```python
country = 'Afghanistan'
country_df = df[df['Entity'] == country]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(country_df['Year'], country_df['Smoking'], marker='o', linestyle='-')
plt.title(f'Trend of Deaths due to Smoking in {country}')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.grid(True)
plt.show()
```

### correlation analysis

```python

numeric_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
correlation_matrix = numeric_df.corr()
```

#### heatmap of correlation matrix

```python
plt.figure(figsize=(8, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Causes of Death')
plt.show()
```

### Comparision between countries

```python
countries = ['Afghanistan','United States', 'India', 'China','Angola','Austria','argentina']
comparison_df = df[df['Entity'].isin(countries)]
```

### Plot comparison of deaths due to air pollution

```python
plt.figure(figsize=(5, 7))
for country in countries:
    subset = comparison_df[comparison_df['Entity'] == country]
    plt.plot(subset['Year'], subset['Air pollution'], marker='o', linestyle='-', label=country)

plt.title('Comparison of Deaths due to Air Pollution')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.legend()
plt.grid(True)
plt.show()
```
